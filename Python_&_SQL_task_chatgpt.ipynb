```python
import pandas as pd
import numpy as np
import plotly.express as px
from faker import Faker
import sqlite3
from datetime import datetime
```

```python
# Initialize Faker
fake = Faker()
np.random.seed(42)

# Number of records
num_records = 10000  # adjust as needed

# Generate data
data = {
    "timestamp": [fake.date_time_between(start_date=datetime(year, 1, 1), end_date=datetime(year, 12, 31))
                  for year in np.random.choice(range(2016,2026), num_records)],
    "threat_type": np.random.choice(["Malware","Phishing","DDoS","Ransomware","Spyware","SQL Injection"], num_records),
    "severity": np.random.choice(["Low","Medium","High","Critical"], num_records, p=[0.2,0.3,0.3,0.2]),
    "source_ip": [fake.ipv4() for _ in range(num_records)],
    "destination_ip": [fake.ipv4() for _ in range(num_records)],
    "country": np.random.choice(["USA","India","China","Germany","UK","Brazil","Russia","Australia"], num_records),
    "detected": np.random.choice([0,1], num_records, p=[0.25,0.75])
}

# Create DataFrame
df = pd.DataFrame(data)
df['year'] = df['timestamp'].dt.year  # add year column for analysis

# Check first rows
df.head()
```
